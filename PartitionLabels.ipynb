### Partition Labels

A string <B>S</B> of lowercase English letters is given. We want to partition this string into as many parts as possible so that each letter appears in at most one part, and return a list of integers representing the size of these parts.

 

Example 1:   

Input: S = "ababcbacadefegdehijhklij"   
Output: [9,7,8]   
Explanation:   
The partition is "ababcbaca", "defegde", "hijhklij".   
This is a partition so that each letter appears in at most one part.   
A partition like "ababcbacadefegde", "hijhklij" is incorrect, because it splits S into less parts.   
 

Note:   

S will have length in range [1, 500].   
S will consist of lowercase English letters ('a' to 'z') only.   

```python
# Solution 1 
# Runtime: 40 ms  (> 78.52%)
# Memory Usage: 13.9 MB (> 49.18%)

class Solution:
    def partitionLabels(self, S: str) -> List[int]:
        
        part = []
        
        '''
        get last index for each element
        '''
        last = {}
        for i in range(len(S)):
            last[S[i]] = i
        
        '''
        get first partition 
        '''
        part.append(last[S[0]] + 1)
        end_idx = last[S[0]]
        stt_idx = -1

        '''
        Update partition
        '''
        for i in range(1, len(S)):
            
            if i > end_idx:
                stt_idx = end_idx
                part.append(last[S[i]] - stt_idx)
                end_idx = last[S[i]]
            else:
                if last[S[i]] > end_idx:
                    part[-1] = last[S[i]] - stt_idx
                    end_idx = last[S[i]]
                
        return part
```

```python
# Solution 2 - Use python buildin string function 'rfind()' to get the last index of each character
# 1st Submission 
# - Runtime: 36 ms (> 91.12%)  / 
# - Memory Usage: 13.7 MB (> 93.05%)  
# 2nd submission 
# - Runtime: 28ms (> 99.29)
# - Memory Usage: 13.8 MB (> 78.73)

class Solution:
    def partitionLabels(self, S: str) -> List[int]:
        
        part = []
        
        '''
        get first substring length
        '''
        part.append(S.rfind(S[0]) + 1)
        end_idx = S.rfind(S[0])
        stt_idx = -1

        '''
        update substring length list
        '''
        for i in range(1, len(S)):
            
            if i > end_idx:
                stt_idx = end_idx
                part.append(S.rfind(S[i]) - stt_idx)
                end_idx = S.rfind(S[i])
            else:
                if S.rfind(S[i]) > end_idx:
                    part[-1] = S.rfind(S[i]) - stt_idx
                    end_idx = S.rfind(S[i])
                
        return part
```

```python
# Solution 3 - use enumerate
# Runtime: 36 ms
# Memory Usage: 14 MB
    
class Solution:
    def partitionLabels(self, S: str) -> List[int]:
        
        last = {ch: i for i, ch in enumerate(S)}
        
        part = []
        stt_idx = end_idx = -1

        for i in range(0, len(S)):
            
            if i > end_idx:
                stt_idx = end_idx
                part.append(last[S[i]] - stt_idx)
                end_idx = last[S[i]]
            else:
                if last[S[i]] > end_idx:
                    part[-1] = last[S[i]] - stt_idx
                    end_idx = last[S[i]]
                
        return part
```

```python
# Sample solution 1 - 20ms 
# Submission test: 40ms/13.8mb

class Solution:
    def partitionLabels(self, S: str) -> List[int]:
        last = {s: i for i, s in enumerate(S)}
        j,anchor = 0,0
        ans = []
        for i, c in enumerate(S):
            j = max(j, last[c])
            if i == j:
                ans.append(i - anchor + 1)
                anchor = i + 1
            
        return ans
```

```python
# Sample solution 2 - 13454kb

class Solution:
    def partitionLabels(self, S: str) -> List[int]:
        last_idx={c:i for i,c in enumerate(S)}
        res=[]
        start,end=0,0
        for i,c in enumerate(S):
            end=max(end,last_idx[c])
            if i==end:
                res.append(end-start+1)
                start=i+1
        return res
```
